```python
%matplotlib widget

import matplotlib.pyplot as plt

#plt.ion()
import numpy as np
import matplotlib
FONT_SIZE = 12
matplotlib.rc('font', size=FONT_SIZE)
matplotlib.rc('axes', titlesize=FONT_SIZE)
import pandas as pd
```

### Get data out of data_sim.csv :

```python
with open("data_sim.csv", "r", encoding="utf-8") as f:
    df = pd.read_csv(f)
    
```

```python
df
```

### Plot Cz vs. Alpha

```python
alpha=df.loc[:,'alpha']
cz=df.loc[:,'CZ']
cx=df.loc[:,'CX']

fig, ax = plt.subplots()
color_orange=np.array([241,105,19])/255
ax.plot(alpha,cz,'ro',markersize=1)
ax.plot(alpha,cx,'b^',markersize=1)
#ax.set_xbound(0,300)
#ax.set_ybound(0,1)
#ax.set_xticks(np.arange(0, 300, 100))
#ax.tick_params(direction='in',pad=6)
#ax.set_yticks(np.arange(0,5,1))
ax.set_xlabel('Angle of attack')
ax.set_ylabel('Cz, Cx')
```

### Plot some moments

```python
fig, ax = plt.subplots()
ax.plot(df.loc[:,'alpha'],df.loc[:,'Cm'],'ro',markersize=1)
ax.set_xlabel('Angle of attack')
ax.set_ylabel('Cm')
# ax.plot(t**2,r2,'b^',markersize=1)
```
